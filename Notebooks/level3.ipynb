```python
import netCDF4 as nc
fname='https://gpm1.gesdisc.eosdis.nasa.gov/opendap/GPM_L3/GPM_3CMB.07/2023/3B-MO.GPM.DPRGMI.CORRAGM.20230101-S000000-E235959.01.V07A.HDF5'
fh=nc.Dataset(fname)
print(fh)
```

```python
var_keys=fh.variables.keys()
csv_f=open('level_3.csv','w')
for key in var_keys:
    ss=fh[key].shape
    dims=','
    for s1 in ss:
        dims+=str(s1)+','
    print(key,dims)
    csv_f.write(key+','+dims[:-1]+'\n')

csv_f.close()
```
